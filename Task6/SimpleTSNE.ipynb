```python
from sklearn import datasets
import json
from openTSNE import TSNE
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
x, y = iris["data"], iris["target"]

embedding = TSNE().fit(x)
```

```python
np.array(embedding).tolist()
```

```python
data = pd.DataFrame(embedding, columns=["Comp1","Comp2"])
data["Label"] = y
```

```python
# TODO add the Target Information into  your json file
with open('TSNE.json', 'w') as f:
    embeddingList = np.array(embedding).tolist();
    yList = y.tolist()
    jsonarray = []
    for i in range(len(y)):
        jsonarray.append({"x": embeddingList[i][0], "y": embeddingList[i][1], "label": yList[i]})
#     print(jsonarray)
    json.dump(jsonarray, f, indent=2)
```

```python
sns.set_context("notebook", font_scale=1.1)
sns.set_style("ticks")

sns.lmplot(x='Comp1',
           y='Comp2',
           data=data,
           fit_reg=False,
           legend=True,
           size=9,
           hue='Label',
           scatter_kws={"s":200, "alpha":0.3}
          )

plt.title("TSNE", weight='bold').set_fontsize('14')
plt.xlabel('Comp 1')
plt.ylabel('Comp 2')
```
